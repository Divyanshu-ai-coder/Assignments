1.Why are dictionary keys required to be immutable in Python?

Explanation:
Python dictionaries use a hashing mechanism to store and retrieve values quickly.
For this to work correctly, the hash value of a key must never change.

Immutable objects (like int, str, tuple) have a fixed hash value.

Mutable objects (like list, set, dict) can change their content, which would change their hash value and break dictionary lookup.

That is why dictionary keys must be immutable.

```python
##example

d = {
    1: "one",
    "name": "Alice",
    (1, 2): "tuple key"
}
print(d[1])          # Output: one
print(d["name"])     # Output: Alice
```

2.Program to create a frequency dictionary from a list

```python
lst = [1, 2, 2, 3, 1, 4, 2]
frequency = {}

for item in lst:
    if item in frequency:
        frequency[item] += 1
    else:
        frequency[item] = 1

print(frequency)
```

3.Difference between dict.get(key) and dict[key]

| Feature        | `dict.get(key)`             | `dict[key]`         |
| -------------- | --------------------------- | ------------------- |
| If key exists  | Returns value               | Returns value       |
| If key missing | Returns `None` (or default) | ❌ Raises `KeyError` |
| Safety         | ✅ Safer                     | ❌ Not safe          |


```python
##example
d = {"a": 10, "b": 20}

print(d.get("c"))        # None
print(d.get("c", 0))     # 0
```

4.Program to merge two dictionaries

```python
d1 = {"a": 10, "b": 20, "c": 30}
d2 = {"b": 25, "c": 15, "d": 40}

merged = {}

for key in d1:
    merged[key] = d1[key]

for key in d2:
    if key in merged:
        merged[key] = max(merged[key], d2[key])
    else:
        merged[key] = d2[key]

print(merged)
```

5.Dictionary comprehension for odd numbers and their cubes

```python
numbers = [1, 2, 3, 4, 5, 6, 7]

cube_dict = {n: n**3 for n in numbers if n % 2 != 0}

print(cube_dict)
```
